```python
import csv

data = [
    ["Nome", "Idade", "Cidade"],
    ["Alice", "30", "São Paulo"],
    ["Bob", "25", "Rio de Janeiro"],
    ["Charlie", "35", "Belo Horizonte"]
]

nome_arquivo = 'pessoas.csv'

with open(nome_arquivo, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(data)
with open(nome_arquivo,'r') as arquivo:
    for linha in arquivo:
        print(linha, end='')
```

```python
import json

dados_json = [
  {"nome": "Alice", "idade": 30, "cidade": "São Paulo"},
  {"nome": "Bob", "idade": 25, "cidade": "Rio de Janeiro"},
  {"nome": "Charlie", "idade": 35, "cidade": "Belo Horizonte"}
]

nome_arquivo = 'dados.json'

with open (nome_arquivo, 'w') as arquivo:
    json.dump(dados_json,arquivo,indent=2)
```
