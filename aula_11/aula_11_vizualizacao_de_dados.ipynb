```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun']
dorama_a = [120, 135, 150, 180, 200, 230]
dorama_b = [80, 85, 90, 100, 110, 125]
dorama_c = [100, 115, 130, 145, 160, 190]
```

```python
plt.plot(meses,dorama_a,label='Dorama A')
plt.xlabel('meses')
plt.ylabel('Popularidade')
plt.title('A popularidade de doramas só tende a crescer em 2024')
plt.plot(meses,dorama_b,label='Dorama B')
plt.plot(meses,dorama_c,label='Dorama C')
plt.legend()
plt.show()
```

```python
plt.plot(meses,dorama_a,color='red',label='Dorama A')
plt.xlabel('meses')
plt.ylabel('Popularidade')
plt.title('A popularidade de doramas só tende a crescer')
plt.plot(meses,dorama_b,label='Dorama B',color='gray',linestyle='dashdot')
plt.plot(meses,dorama_c,label='Dorama C',color='gray',linestyle='--', marker='o',markersize=5,markerfacecolor='gray')
plt.legend()
plt.show()
```

```python
import numpy as np
```

```python
t = np.linspace(0,10,100)
x = np.sin(t)
y = np.cos(t)

plt.plot(t,x,label='Seno de t',color='blue')
plt.plot(t,y,label='Seno de t',color='red')
plt.legend()
plt.show()
```

```python
t = np.linspace(0,10,100)
x = np.sin(t)
y = np.cos(t)
z = np.tan(t)

plt.plot(t,x,label='Seno de t',color='blue')
plt.plot(t,y,label='Cosseno de t',color='red')
plt.plot(t,z,label='Tangente de t',color='orange')
plt.legend()
plt.show()
```

```python
dados = {
    'Gênero': ['Romance', 'Comédia', 'Drama', 'Fantasia', 'Mistério'] * 20,
    'Avaliação': np.random.normal(8, 0.5, 100)
}

df = pd.DataFrame(dados)

df
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Gênero',y='Avaliação',data=df)
plt.title('Distribuição de Avaliações por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Avaliação')
plt.show()
```

```python
fig,axs  = plt.subplots(2,2,figsize=(10,10))
#gráfico 1 hexbin
axs[0,0].hexbin(num_episodios,avaliacoes,gridsize=20,cmap='Blues')
axs[0,0].set_title('Distribuição de Avaliações')
axs[0,0].set_xlabel('Número de Episódios')
axs[0,0].set_ylabel('Avaliações')

#gráfico 2 boxplot
sns.boxplot(x='Gênero',y='Avaliação',data=df,ax=axs[0,1])
axs[0, 1].set_title('Boxplot: Avaliações por Gênero')

#gráfico 3: displot
sns.histplot(avaliacoes,kde=True,ax=axs[1,0])
axs[1,0].set_title('Histograma: Distribuição de Avaliações')
axs[1,0].set_xlabel('Avaliações')

#gráfico 4: gráfico de barras
anos = ['2018', '2019', '2020', '2021', '2022']
quantidade_doramas = [15, 20, 25, 30, 22]
axs[1,1].bar(anos,quantidade_doramas,color='teal')
axs[1,1].set_title('Quantidade de Doramas por Ano')
axs[1,1].set_xlabel('Ano')
axs[1,1].set_ylabel('Quantidade de Doramas')

plt.show()
```
